```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("/content/train.csv", index_col=0)
test = pd.read_csv("/content/test.csv", index_col=0).drop(columns=["feature_5"])
```

```python
X = train.drop(columns=["feature_5", "target"])
y = train["target"]
test = test
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

oof_train = np.zeros(len(X))
test_predictions = np.zeros(len(test))

for train_index, val_index in kf.split(X, y):
    X_train_fold = X.iloc[train_index]
    y_train_fold = y.iloc[train_index]

    X_val_fold = X.iloc[val_index]
    y_val_fold = y.iloc[val_index]

    model = LogisticRegression()
    model.fit(X_train_fold, y_train_fold)

    pred_val = model.predict_proba(X_val_fold)
    pred_val = pred_val[:, 1]

    oof_train[val_index] = pred_val

    pred_test = model.predict_proba(test)[:, 1]

    test_predictions = test_predictions + pred_test
```

```python
test_predictions = test_predictions / 5
test_predictions
```

```python
score_oof = roc_auc_score(y, oof_train)
print(score_oof)
```

```python
submission = pd.DataFrame({"id": test.index, "Heart Disease": test_predictions})

submission.to_csv("submission.csv", index=False)
print("Submission saved!")
```
